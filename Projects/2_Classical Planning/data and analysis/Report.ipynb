```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('Results.csv')
df
```

```python
exp_dict = {20: 0, 72: 0, 88: 0, 104: 0}
problems = ['One\n(20 Actions)', 'Two\n(72 Actions)', 'Three\n(88 Actions)', 'Four\n(104 Actions)']
actions = df['Actions']
expansions = df['Expansions']
plt.figure(figsize = (8, 4))
  
for i in range(len(actions)):
    exp_dict[actions[i]] = exp_dict[actions[i]] + expansions[i]
        
print(exp_dict)        

plt.bar(range(len(exp_dict)), exp_dict.values(), align = 'center');
plt.xticks(range(4), problems);

plt.xlabel('Problem Number', color = 'blue', weight = 800, style = 'oblique');
plt.ylabel('Sum of Number of Expansions\nfor All Algorithms Attempted', color = 'blue', weight = 800, style = 'oblique');
```

```python
time_dict = {20: 0, 72: 0, 88: 0, 104: 0}
problems = ['One\n(20 Actions)', 'Two\n(72 Actions)', 'Three\n(88 Actions)', 'Four\n(104 Actions)']
actions = df['Actions']
times = df['Time in seconds']
plt.figure(figsize = (8, 4))
  
for i in range(len(actions)):
    time_dict[actions[i]] = time_dict[actions[i]] + times[i]
        
print(time_dict)        

plt.bar(range(len(time_dict)), time_dict.values(), align = 'center');
plt.xticks(range(4), problems);

plt.xlabel('Problem Number', color = 'blue', weight = 800, style = 'oblique');
plt.ylabel('Sum of Time In Seconds\nfor All Algorithms Attempted', color = 'blue', weight = 800, style = 'oblique');
```

```python
path_dict = {20: 0, 72: 0, 88: 0, 104: 0}
problems = ['One\n(20 Actions)', 'Two\n(72 Actions)', 'Three\n(88 Actions)', 'Four\n(104 Actions)']
actions = df['Actions']
path_length = df['Path Length']
plt.figure(figsize = (8, 4))
  
for i in range(len(actions)):
    path_dict[actions[i]] = path_dict[actions[i]] + path_length[i]
        
print(path_dict)        

plt.bar(range(len(path_dict)), path_dict.values(), align = 'center');
plt.xticks(range(4), problems);

plt.xlabel('Problem Number', color = 'blue', weight = 800, style = 'oblique');
plt.ylabel('Sum of Length of Path\nfor All Algorithms Attempted', color = 'blue', weight = 800, style = 'oblique');
```

```python
path_dict = {20: 0, 72: 0, 88: 0, 104: 0}
problems = ['One\n(20 Actions)', 'Two\n(72 Actions)', 'Three\n(88 Actions)', 'Four\n(104 Actions)']
actions = df['Actions']
path_length = df['Path Length']
plt.figure(figsize = (8, 4))
  
for i in range(len(actions)):
    if i % 11 != 1:
        path_dict[actions[i]] = path_dict[actions[i]] + path_length[i]
        
print(path_dict)        

plt.bar(range(len(path_dict)), path_dict.values(), align = 'center');
plt.xticks(range(4), problems);

plt.xlabel('Problem Number', color = 'blue', weight = 800, style = 'oblique');
plt.ylabel('Sum of Length of Path\nfor All Algorithms Attempted\nExcept Depth First Search', color = 'blue', weight = 800, style = 'oblique');
```
